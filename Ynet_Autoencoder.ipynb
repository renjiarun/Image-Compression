```python
import numpy as np
from tensorflow.keras.layers import Dense, Dropout, Input, Conv2D, MaxPooling2D,UpSampling2D, Flatten, concatenate, Reshape
from tensorflow.keras.models import Model
import cv2
import matplotlib.pyplot as plt
from skimage.metrics import structural_similarity as compare_ssim
from PIL import Image

# Load the color image (replace 'path/to/your/image.jpg' with your image path)
image = cv2.imread('/content/gdrive/MyDrive/jpg_Set5/kodim07 (1).jpg')

print("Color Image Input!")
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()

# Resize the image to 256x256
resized_image = cv2.resize(image, (256, 256), interpolation=cv2.INTER_AREA)

print("Color Image Resized!")
plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
plt.show()

# Save the resized image
cv2.imwrite('/content/gdrive/MyDrive/jpg_Set5/kodim07_256_Colour.jpg', resized_image)

# Prepare the image for the neural network
OrgImg = resized_image
image1 = resized_image

# Normalize the image between 0 and 1
image1 = resized_image / 255.0

# Convert to float32 to avoid unsupported depth error
image1 = image1.astype(np.float32)

# Add batch and channel dimensions (batch_size, height, width, channels)
image1 = image1.reshape((1, 256, 256, 3))

# Convert BGR to RGB explicitly (if not already in RGB)
image_rgb = cv2.cvtColor(image1[0], cv2.COLOR_BGR2RGB)

print("Image Input to Y-Network!")
plt.imshow(image_rgb)
plt.show()
print(image_rgb)
#plt.imshow(cv2.cvtColor(image1, cv2.COLOR_BGR2RGB))
#plt.show()

# Convert the reconstructed image back to the range [0, 255]
image_rgb = (image_rgb * 255).astype(np.uint8)
plt.imshow(image_rgb)
plt.show()
print(image_rgb)


# Network parameters
input_shape = (256, 256, 3)  # Updated for 3 channels (RGB)
batch_size = 1
kernel_size = 3
dropout = 0.0
n_filters = 32

# Left branch of Y network
left_inputs = Input(shape=input_shape)
x = left_inputs
filters = n_filters

# 3 layers of Conv2D-Dropout-MaxPooling2D
# number of filters doubles after each layer (32-64-128)
for i in range(3):
    x = Conv2D(filters=filters, kernel_size=kernel_size, padding='same', activation='relu')(x)
    x = Dropout(dropout)(x)
    x = MaxPooling2D()(x)
    filters *= 2

# Right branch of Y network
right_inputs = Input(shape=input_shape)
y = right_inputs
filters = n_filters

# 3 layers of Conv2D-Dropout-MaxPooling2D
# number of filters doubles after each layer (32-64-128)
for i in range(3):
    y = Conv2D(filters=filters, kernel_size=kernel_size, padding='same', activation='relu', dilation_rate=2)(y)
    y = Dropout(dropout)(y)
    y = MaxPooling2D()(y)
    filters *= 2
print(x)
print(y)
# Merge left and right branches outputs
y = concatenate([x, y])
print(y)
# Feature maps to vector before connecting to Dense


# Now, we need to adjust the UpSampling2D layers to expand the feature maps to 512x512

# Apply UpSampling and Conv2D layers to increase spatial dimensions
for i in range(4):  # Adjust the number of layers and filters as needed
    y = UpSampling2D()(y)
    y = Conv2D(filters=filters // 2, kernel_size=kernel_size, padding='same', activation='relu')(y)
    filters //= 2  # Reduce the number of filters

# Final Conv2D layer to ensure output has 3 channels (RGB)
output_img = Conv2D(3, kernel_size=(3, 3), padding='same', activation='sigmoid')(y)

# Create and compile the model
model = Model(inputs=[left_inputs, right_inputs], outputs=output_img)
model.compile(optimizer='adam', loss='mse')

# Print the model summary
model.summary()

y = Flatten()(y)
y = Dropout(dropout)(y)

print(y.shape[1])
# Calculate the total number of elements after flattening
flattened_dim = np.prod(y.shape[1:])
print(f"Flattened dimension: {flattened_dim}")

# To calculate the flattened dimension
model = Model(inputs=[left_inputs, right_inputs], outputs=y)
flattened_dim = model.output_shape[-1]
print(f"Flattened dimension 2: {flattened_dim}")

# Autoencoder
encoded_dim = 128  # Dimension of the compressed representation
encoded = Dense(encoded_dim, activation='relu')(y)
print(f"End of encoder: {encoded}")

# Decoder
decoded = Dense(np.prod(y.shape[1:]), activation='relu')(encoded)
decoded = Reshape((32, 32, 256))(decoded)
print(f"End of decoder: {decoded}")

# Define the autoencoder model separately
autoencoder = Model(inputs=encoded, outputs=decoded)

# Compile the autoencoder model
autoencoder.compile(optimizer='adam', loss='mse')

# Print the autoencoder model summary
autoencoder.summary()

# Save the autoencoder model separately
autoencoder.save('autoencoder_Ynet_model_color.h5')
```
